```python
def prompt_classify(question_ia, answer_ia):
    content_string = f'Eres un experto en analizar subjetividades politicas. Basado en la siguiente pregunta {question_ia} y en su respectiva respuesta {answer_ia} clasifica a cual categoria corresponde POLARIZACIÓN, EMOCIONES POLÍTICAS, IDENTIDAD. Para entender mejor el contexto de la pregunta y la respuesta, se proporciona el siguiente contexto: "POLARIZACIÓN: Es un proceso psicosocial en el que las perspectivas sobre un problema específico se reducen a dos posturas extremadamente opuestas y mutuamente excluyentes dentro de una sociedad determinada. Fenómeno que emerge en sociedades caracterizadas por divisiones de clases sociales. Es un proceso en el que se desarrolla una elaboración ideológica de intereses sociales, lo que da lugar a la formación de grupos que tienen intereses particulares (Martín-Baró, 2003).", "EMOCIONES POLÍTICAS: Emociones que tienen un contenido político, o que están relacionados a ello. Estas emociones motivan a las personas a tomar acción política, a conectarse con otros que sienten las mismas necesidades o preocupaciones, y a desconectarse de aquellas personas con las que no comparten emociones. Estas son necesarias para la democracia, ya que permiten que las personas se comprometan y trabajen por objetivos comunes. (Nussbaum, 2013)." y "IDENTIDAD: Según Goffman (2006), la identidad es el conjunto de rasgos y características que definen a una persona o un grupo social, y que se expresan a través de la forma de presentarse y relacionarse con los demás.". Realiza la clasificacion. Solo dame el nombre de la categoria a la que pertenece la pregunta. No agregues informacion adicional.'
    # Simulated API call to the language model
    response = client.chat.completions.create(
        model="gpt-4-turbo-preview",
        messages=[{"role":"user", "content":content_string}],
        temperature=0
    )
    return response.choices[0].message.content
```

```python
i=43
question = preguntas.Preguntas.to_list()
print(question[i])
columnas = respuestas.columns.to_list()
print(columnas[i+1])

respuestas[f'Prediccion a la pregunta {columnas[i+1]}']= ''
for j in range(respuestas.shape[0]):
    print(f'Pregunta: {question[i]}')
    print(f'Respuesta: {respuestas[columnas[i+1]][j]}')
    respuesta_ia = prompt_classify(question[i],respuestas[columnas[i+1]][j])
    respuestas[f'Prediccion a la pregunta {columnas[i+1]}'][j] = respuesta_ia
    print('\n')
```

```python
def prompt_classify_emotion(question_ia, answer_ia):
    content_string = f'Eres un experto en analizar subjetividades politicas. Basado en la siguiente pregunta {question_ia} y en su respectiva respuesta {answer_ia}. La categoria es EMOCIONES POLÍTICAS. clasifica a cual subcategoria corresponde PRIMARIA, SECUNDARIA. Para entender mejor el contexto de la pregunta y la respuesta, se proporciona el siguiente contexto: "PRIMARIA: Son aquellas innatas o universales que desde Nussbaum (2013), se pueden definir como emociones de miedo que pueden llevar a las personas a protegerse a sí mismos o sus seres queridos, Las emociones de vergüenza y desprecio se diferencian entre ellas de manera que la primera puede llevar a las personas a luchar contra injusticias, y la segunda puede justificar la discriminación. Y las emociones de compasión y amor son aquellas que motivan a trabajar para construir comunidades más solidarias." y "SECUNDARIA: Son aquellas aprendidas o culturales como el patriotismo o nacionalismo (Dahl, 1989).". Realiza la clasificacion. Solo dame el nombre de la subcategoria a la que pertenece la pregunta. No agregues informacion adicional.'
    # Simulated API call to the language model
    response = client.chat.completions.create(
        model="gpt-4-turbo-preview",
        messages=[{"role":"user", "content":content_string}],
        temperature=0
    )
    return response.choices[0].message.content

def prompt_classify_identity(question_ia, answer_ia):
    content_string = f'Eres un experto en analizar subjetividades politicas. Basado en la siguiente pregunta {question_ia} y en su respectiva respuesta {answer_ia}. La categoria es IDENTIDAD. clasifica a cual subcategoria corresponde SOCIAL, INDIVIDUAL. Para entender mejor el contexto de la pregunta y la respuesta, se proporciona el siguiente contexto: "SOCIAL: La identidad social es la parte de nuestra identidad que viene de los grupos a los que pertenecemos (Rincón-Villanueva,2020). Esta se deriva de la pertenencia a un grupo social, junto con su significado emocional y valorativo asociado. Tajfel (1984)." y "INDIVIDUAL: Según las contribuciones teóricas de Hale (2004), la identidad personal se puede ver como un conjunto de elementos que están conectados entre sí, los cuales pueden cambiar con el tiempo en términos de lo que significan y cuán importantes son. Algunos de estos elementos son la edad, el género, la profesión, la religión y la nacionalidad, entre otros.". Realiza la clasificacion. Solo dame el nombre de la subcategoria a la que pertenece la pregunta. No agregues informacion adicional.'
    # Simulated API call to the language model
    response = client.chat.completions.create(
        model="gpt-4-turbo-preview",
        messages=[{"role":"user", "content":content_string}],
        temperature=0
    )
    return response.choices[0].message.content
```
